# Importing Required Libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('Big Mart Sales Prediction - Preprocessing.xlsx',index_col=0)
```

```python
df.boxplot(column='Item_Outlet_Sales',by='Item_Fat_Content')
plt.figure(figsize=(15,10))
```

##### Here, We can say that Some of ‘Low Fat’ values mis-coded as ‘low fat’ and ‘LF’. Also, some of ‘Regular’ are mentioned as ‘regular’.

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({'LF':'Low Fat', 'reg':'Regular', 'low fat':'Low Fat'})
print('After Replacing: \n',df['Item_Fat_Content'].value_counts())
```

# Exploratory Data Analysis

### Count Plot of Item_Fat_Content

```python
sns.countplot(df["Item_Fat_Content"] , palette="Set3")
```

### Count Plot of Item_Type

```python
l = list(df['Item_Type'].unique())
chart = sns.countplot(df["Item_Type"],palette="Set3")
chart.set_xticklabels(labels=l, rotation=90)
```

#### From the above plot, Snack Food & Fruits and Vegetables Item_Type have the maximum count. Seafood have the lowest count.

### Count Plot of Outlet_Establishment_Year

```python
sns.countplot(df['Outlet_Establishment_Year'] , palette="Set3")
```

#### In the year 1985 Maximum number of outlets are established and in 1998 Minimum number of Outlets are established.

### Count Plot of Outlet_Size

```python
sns.countplot(df['Outlet_Size'], palette="Set3")
```

#### Medium sized outlets are large in number and High sized outlets are small in number.

### Count Plot of Outlet_Type

```python
f, axs = plt.subplots(figsize=(8,8))
sns.countplot(df['Outlet_Type'] , palette = 'Set3')
```
